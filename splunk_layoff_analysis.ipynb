# Rename column names for easy access

```python
data = data.rename(columns={
    'xSplunkers, add your info here!! https://forms.gle/AAryt8nDnNvHbwZy5': 'Timestamp',
    'Unnamed: 1': 'First Name',
    'Unnamed: 2': 'Last Name',
    'Unnamed: 3': 'Former Job Title',
    'Unnamed: 4': 'Team / Function',
    'If you\'ve gotten a new job and want to be crossed off the list, email splunk@edrabbit.com': 'City',
    'Unnamed: 6': 'Remote',
    'Unnamed: 7': 'LinkedIn URL',
    'Unnamed: 8': 'comments'
})
```

# Filtering

**Next Step:** filter out non-engineers, frontend engineers, and engineers who must work from home (Samsung requires engineers to go office

```python
onsite_eng = data[
    (data['Former Job Title'].isna() | data['Former Job Title'].str.contains('engineer', case=False)) &
    (data['Former Job Title'].isna() | ~data['Former Job Title'].str.contains('frontend', case=False)) &
    (data['Remote'].isna() | data['Remote'].str.contains('hybrid', case=False) | data['Remote'].str.contains('office', case=False))
    ]
print(onsite_eng.shape)
print(onsite_eng.head())
```

**Next Step**: we seems still have 54 engineers; let us narrow down to engineers in Silicon Valley and california

```python
ca_onsite_eng = onsite_eng[ onsite_eng['City'].isna() | 
                            onsite_eng['City'].str.contains('CA', case=False) |
                            onsite_eng['City'].str.contains('San Francisco', case=False) |
                            onsite_eng['City'].str.contains('San Mateo', case=False) |
                            onsite_eng['City'].str.contains('Santa Clara', case=False)
                           ]
print(ca_onsite_eng.shape)
print(ca_onsite_eng.head())

print(ca_onsite_eng['LinkedIn URL'])
```

**Next Step**: we still have 22 engineers; maybe we should go through these 22 engineers manually. But let us identify data engineers and prioritize these engineers

```python
data_ca_onsite_eng = ca_onsite_eng[
    ca_onsite_eng['Former Job Title'].isna() |
    ca_onsite_eng['Former Job Title'].str.contains('data', case=False) |
    ca_onsite_eng['Team / Function'].isna() |
    ca_onsite_eng['Team / Function'].str.contains('data', case=False)
]

print(data_ca_onsite_eng.shape)
print(data_ca_onsite_eng.head())
print(data_ca_onsite_eng['LinkedIn URL'])
```
